# Q1. Which of the following is mutable in nature Series, DataFrame, Panel?




In pandas, both Series and DataFrame are mutable in nature, meaning their values can be changed after they are created.

A Series is a one-dimensional array-like object that can hold any data type, such as integers, floats, strings, and Python objects. You can change the values in a Series by assigning new values to specific index positions.

A DataFrame is a two-dimensional tabular data structure consisting of rows and columns, similar to a spreadsheet or SQL table. You can modify the values in a DataFrame by assigning new values to specific cells, columns, or rows.

On the other hand, a Panel is a three-dimensional data structure, which is a collection of multiple DataFrames. In pandas version 1.0.0 and later, the Panel is deprecated and replaced with the more efficient and flexible multi-index DataFrame. Multi-index DataFrame is also mutable like DataFrame.


# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional tabular data structure consisting of rows and columns, similar to a spreadsheet or SQL table. It is the most commonly used Pandas object and is designed to handle a variety of data types, including numerical, categorical, and textual data.

On the other hand, a Series is a one-dimensional labeled array that can hold data of any type, such as integers, floats, strings, or Python objects. Each element in a Series has a label called an index.

Here's an example to illustrate the difference between a DataFrame and a Series:

```python
import pandas as pd

# create a Series
s = pd.Series([1, 2, 3, 4, 5], name='Numbers')

# create a DataFrame
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['New York', 'San Francisco', 'Seattle']})

# print the Series and DataFrame
print(s)
print(df)
```

In the code, we first create a Series called s that contains five integers and is labeled as 'Numbers'. We then create a DataFrame called df that contains three columns ('Name', 'Age', and 'City') and three rows, with each row representing the data for one person.

The main differences between a Series and a DataFrame are:

A Series has only one column of data, while a DataFrame can have multiple columns.
A Series has only one index, while a DataFrame can have multiple row and column indexes.
A DataFrame is a two-dimensional object, while a Series is a one-dimensional object.
In the output, you can see that the Series s has only one column and one index, while the DataFrame df has three columns and three rows, with both row and column indexes.
